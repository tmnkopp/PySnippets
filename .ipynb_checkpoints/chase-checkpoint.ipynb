```python
dfd = pd.read_csv('data/d.csv')# dfd dataframe dirty, lets clean it 
dfd = dfd.iloc[:,[1,2,3,5]] # filter only colums we need
dfd = dfd[dfd['Amount'] < 0] # filter amount by expenses ex. -10
dfd = dfd.rename(columns = {'Post Date':'Date'})
dfd['Date'] = pd.to_datetime(dfd['Date']) # convert to datetime datatye
dfd['Amount'] = dfd['Amount'].abs() # convert -10 to + 10
```
